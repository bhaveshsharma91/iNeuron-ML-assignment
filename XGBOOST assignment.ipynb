```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('adult.data',header = None)
```

```python
test = pd.read_csv('adult.test', skiprows = 1, header = None)
```

```python
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation','relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'wage_class'] 
 
train.columns = col_labels 
 
test.columns = col_labels
```

```python
df = pd.concat([train,test])
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(col,df[col].unique(), end="\n\n")
```

#### # these columns have " ?" which are missing values


```python

df.workclass = df.workclass.replace(" ?",np.nan)
df.occupation = df.occupation.replace(" ?",np.nan)
df.native_country = df.native_country.replace(" ?",np.nan)
```

```python
df.info()
```

```python
df.dropna(inplace=True) # drop those rows as it is not possible to predict them
```

```python
df.drop(columns=['capital_loss','capital_gain'],inplace=True) # these columns have too many missing values so drop them
```

```python
df.head()
```

```python
for i in df['education_num'].unique():
    print(i,df[df["education_num"]==i]['education'].unique())
```

```python
# we can drop education column...
df.drop(columns=['education'],inplace=True)
```

```python
df.head()
```

```python
# make wage_class into numeric
df.wage_class = df.wage_class.replace({' <=50K.': 0,' >50K.':1,' <=50K': 0,' >50K':1})
```

```python
df.wage_class
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col]= df[col].astype('category')
        df[col] = df[col].cat.codes
```

```python
# scaling our values

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.iloc[:,:-1])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled,df.iloc[:,-1],test_size=0.3)
```

```python
from xgboost import XGBClassifier

clf = XGBClassifier(objective='binary:logistic')
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

### 82.8%

```python
clf.score(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
   
    ' learning_rate':[1,0.5,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
    
}
```

```python
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
new_model=XGBClassifier(learning_rate= 1, max_depth= 3, n_estimators= 50)
new_model.fit(X_train, y_train)
```

```python
new_model.score(X_test,y_test)
```

### 82.7%

### Performance metrics

```python
y_pred = new_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score, accuracy_score

confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```
